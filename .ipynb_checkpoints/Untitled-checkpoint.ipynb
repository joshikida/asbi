## Oracle function

here is the oracle function (and corresponding inference pipeline \[placeholder]) which is called for selected instrument design(s). in the real world this would be quite expensive to run

```python
def inference_model(noisy_img: np.ndarray) -> norm:
    """
    placeholder for the analysis pipeline for parameter inference.  
    returns an inferred posterior for theta_E.
    higher noise in the image results in a wider (more uncertain) posterior.
    """
    # simulate the effect of noise on inference quality
    image_noise_level = np.std(noisy_img - img)
    
    # posterior's uncertainty increases with the image noise
    base_uncertainty = 0.005
    posterior_std = base_uncertainty + (image_noise_level * 0.2)
    
    #  posterior mean will be close to the true value, with some noise-dependent bias
    posterior_mean = TRUE_THETA_E + (np.random.randn() * image_noise_level * 0.05)
    
    return norm(loc=posterior_mean, scale=posterior_std)

def oracle(design_param_x: float) -> float:
    """
    simulation and analysis pipeline. takes the instrument design parameter (noise_std) and returns its performance.
    """
    #simulate an observation for the given instrument design
    noise = np.random.normal(0, design_param_x, img.shape)
    observed_image = img + noise
    
    #perform the "science analysis"  inference
    inferred_posterior = inference_model(observed_image)
    
    #performance metric is the uncertainty (std dev) of the posterior. goal is to minimize
    performance_metric_y = inferred_posterior.std()
    
    return performance_metric_y
```

```python
def oracle(design_param_x: float) -> float:
    """
    A more complex oracle with a non-linear "sweet spot".
    """
    # a. Simulate the observation
    noise = np.random.normal(0, design_param_x, img.shape)
    observed_image = img + noise
    
    # b. Perform science analysis
    inferred_posterior = inference_model(observed_image)
    
    # c. Calculate the performance metric
    performance_from_inference = inferred_posterior.std()
    
    # d. NEW: Add a penalty for being far from the "sweet spot"
    optimal_noise = 0.1
    # This quadratic term creates the U-shape
    design_penalty = 0.5 * (design_param_x - optimal_noise)**2 
    
    # The final performance is a combination of inference quality and design penalty
    final_performance_y = performance_from_inference + design_penalty
    
    return final_performance_y[0]
```

```python
def oracle(design_param_x: float) -> float:
    """
    An oracle with a "performance cliff".
    """
    # NEW: Check if the design is past the cliff
    if design_param_x > 0.3:
        # Performance is terrible past this point
        return 1.0 
    
    # Otherwise, calculate performance as normal
    noise = np.random.normal(0, design_param_x, img.shape)
    observed_image = img + noise
    inferred_posterior = inference_model(observed_image)
    performance_metric_y = inferred_posterior.std()
    
    return performance_metric_y
```

```python
def plot_results(X_train, y_train,inptitle):
    """
    Visualize the results of the active learning process.
    """
    print("\n--- Generating final plot ---")
    
    # For visualization, let's find the "true" performance landscape
    # In a real scenario, this is what we are trying to discover
    true_y = [oracle(x) for x in x_space]

    # Train a final surrogate on all queried points for plotting
    kernel = 1.0 * RBF(length_scale=1.0) + WhiteKernel(noise_level=1e-5)
    final_gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
    final_gp.fit(X_train, y_train)
    y_pred, sigma = final_gp.predict(x_space, return_std=True)

    plt.figure(figsize=(12, 8))
    # Plot the true, unknown function
    plt.plot(x_space, true_y, 'r--', label='True Performance (Unknown to Algorithm)')
    
    # Plot the GP surrogate's final prediction and confidence interval
    plt.plot(x_space, y_pred, 'b-', label='Surrogate Model Prediction (GP)')
    plt.fill_between(x_space.ravel(), y_pred - 1.96 * sigma, y_pred + 1.96 * sigma,
                     alpha=0.2, color='blue', label='95% Confidence Interval')
                     
    # Plot the points that were actually queried by the active learner
    plt.scatter(X_train, y_train, c='black', s=100, zorder=10, edgecolor='white', label='Queried Design Points')
    
    # Find and highlight the best design found
    best_index = np.argmin(y_train)
    best_x, best_y = X_train[best_index], y_train[best_index]
    plt.scatter(best_x, best_y, c='gold', s=200, zorder=11, edgecolor='black', marker='*', label=f'Best Design Found')
    
    plt.title(inptitle, fontsize=16)
    plt.xlabel("Instrument Design Parameter (Noise Std. Dev.)", fontsize=12)
    plt.ylabel("Performance Metric (Posterior Uncertainty on $\\theta_E$)", fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(True, linestyle=':')
    plt.show()
```

## Active SBI Loop

This is the main active sbi loop. 

```python
def acqui_ucb (y_pred,sigma,y_train):
    # we want to MINIMIZE y, so we look for points with low predicted y (low -y_pred) and high uncertainty (high sigma).
    # y_pred = surrogate model's best guess for the performance, sigma = unceratinty, beta = tradeoff
    beta = 1.96 # tradeoff parameter
    ucb_scores = -y_pred + beta * sigma # calculate utility score 
    # take the max "utilitous" design and extract its parameter set
    next_design_index = np.argmax(ucb_scores)
    next_design_x = x_space[next_design_index]

    return next_design_x
```

```python
def acqui_random (y_pred,sigma,y_train):

    next_design_index = np.random.choice(len(x_space))        
    next_design_x = x_space[next_design_index]
    return next_design_x
```

```python
def acqui_greedy(y_pred, sigma,y_train):
    """
    Selects the design with the best predicted performance, ignoring uncertainty.
    This is a pure exploitation strategy.
    """
    # We want to MINIMIZE y, so we find the index of the minimum predicted value.
    next_design_index = np.argmin(y_pred)
    next_design_x = x_space[next_design_index]

    return next_design_x
```

```python
def acqui_uncertainty(y_pred, sigma,y_train):
    """
    Selects the design where the surrogate model is most uncertain.
    This is a pure exploration strategy.
    """
    # Find the index of the maximum uncertainty (sigma).
    next_design_index = np.argmax(sigma)
    next_design_x = x_space[next_design_index]

    return next_design_x
```

```python
def acqui_ei(y_pred, sigma, y_train):
    """
    Selects the design with the highest Expected Improvement (EI).
     hybrid strategy.
    """
    # Find the best performance seen so far
    best_y_so_far = np.min(y_train)
    
    # Ensure sigma is not zero to avoid division by zero errors
    sigma = np.maximum(sigma, 1e-9)
    
    # Calculate the improvement over the best point
    improvement = best_y_so_far - y_pred
    
    # Z-score: standardized improvement
    Z = improvement / sigma
    
    # Calculate the Expected Improvement score for each point
    ei_scores = improvement * norm.cdf(Z) + sigma * norm.pdf(Z)
    
    # Select the design with the maximum EI score
    next_design_index = np.argmax(ei_scores)
    next_design_x = x_space[next_design_index]

    return next_design_x
```

```python
def policy_ucb(y_pred, sigma, x_space, y_train):
    beta = 1.96
    ucb_scores = -y_pred + beta * sigma
    return x_space[np.argmax(ucb_scores)]

def policy_random(y_pred, sigma, x_space, y_train):
    return np.random.choice(x_space.ravel())

def policy_ei(y_pred, sigma, x_space, y_train):
    best_y_so_far = np.min(y_train)
    sigma = np.maximum(sigma, 1e-9)
    improvement = best_y_so_far - y_pred
    Z = improvement / sigma
    ei_scores = improvement * norm.cdf(Z) + sigma * norm.pdf(Z)
    return x_space[np.argmax(ei_scores)]

# --- EXECUTION ---
# Create an instance of the ActiveLearner, passing in the policy you want to test
learner = ActiveLearner(x_space=x_space, acqui_policy=policy_ucb)

# Run the experiment with live plotting enabled
queried_X, queried_y = learner.run(total_queries=20, plot_progress=True)
```

```python
def active_learning_loop(acquisition_func, initial_points=3, total_queries=15):
    """
    main Active SBI loop.
    """
    print(f"\n--- starting active learning ---")
    print(f"perofrming {initial_points} initial random queries for cold start...")
    
    # initialization (cold start). take some random points in the instrument parameter space and oracle them
    X_train = []
    y_train = []
    initial_indices = np.random.choice(len(x_space), initial_points, replace=False)
    for i in initial_indices:
        design_x = x_space[i]
        performance_y = oracle(design_x)
        X_train.append(design_x)
        y_train.append(performance_y)

    # main loop 
    for i in range(initial_points, total_queries):
        print(f"\nIteration {i+1}/{total_queries}")
        
        ## train the surrogate (gaussian process) on training data thus far
        kernel = 1.0 * RBF(length_scale=1.0) + WhiteKernel(noise_level=1e-5)
        gp_surrogate = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
        gp_surrogate.fit(np.array(X_train), np.array(y_train))
        
        ## select next design with Upper Confidence Bound (UCB) acquisition function
        y_pred, sigma = gp_surrogate.predict(x_space, return_std=True)
        next_design_x = acquisition_func(y_pred,sigma,y_train)
        
        print(f"  Surrogate trained on {len(X_train)} points.")
        print(f"  Acquisition policy selected next design: noise_std = {next_design_x[0]:.3f}")

        ## oracle on selected design
        new_y = oracle(next_design_x)
        print(f"  Oracle returned performance (uncertainty): {new_y:.4f}")
        
        ## add to training data
        X_train.append(next_design_x)
        y_train.append(new_y)

    return np.array(X_train), np.array(y_train)
```
